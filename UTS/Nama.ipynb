```python
import cv2
import numpy as np
import matplotlib.pyplot as plt # Dipindahkan ke atas agar lebih rapi

def find_threshold(image_path):
    """
    Mencari ambang batas (threshold) yang optimal untuk gambar JPG menggunakan metode Otsu.

    Args:
        image_path (str): Path ke file gambar JPG.

    Returns:
        tuple: Tuple berisi (ambang_batas, gambar_biner).
               - ambang_batas (float): Nilai ambang batas yang ditemukan oleh metode Otsu.
               - gambar_biner (numpy.ndarray): Gambar biner hasil thresholding.
               Mengembalikan None jika gambar tidak dapat dibaca.
    """
    try:
        # Baca gambar dalam skala abu-abu
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Error: Tidak dapat membaca gambar dari path: {image_path}")
            return None

        # Aplikasikan Gaussian blur untuk mengurangi noise (opsional tapi sering membantu)
        blur = cv2.GaussianBlur(img, (5, 5), 0)

        # Gunakan metode Otsu untuk mencari ambang batas optimal
        _, thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        ret, _ = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        return ret, thresh
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
        return None

if __name__ == "__main__":
    file_path = 'halo.jpg'  # Sekarang kita langsung menggunakan 'halo.jpg'

    try:
        # Coba baca gambar untuk memastikan file ada
        img_check = cv2.imread(file_path)
        if img_check is None:
            print(f"Error: File gambar '{file_path}' tidak ditemukan atau tidak dapat dibaca.")
        else:
            result = find_threshold(file_path)
            if result:
                threshold_value, binary_image = result
                print(f"Ambang batas yang ditemukan untuk '{file_path}': {threshold_value}")

                # Tampilkan gambar asli dan gambar biner
                plt.figure(figsize=(10, 5))

                plt.subplot(1, 2, 1)
                original_image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                plt.imshow(original_image, cmap='gray')
                plt.title('Gambar Asli (Grayscale)')
                plt.axis('off')

                plt.subplot(1, 2, 2)
                plt.imshow(binary_image, cmap='gray')
                plt.title(f'Gambar Biner (Threshold: {threshold_value})')
                plt.axis('off')

                plt.tight_layout()
                plt.show()
            else:
                print(f"Gagal menemukan ambang batas untuk '{file_path}'.")
    except ImportError:
        print("Error: Library matplotlib tidak terinstal. Tidak dapat menampilkan gambar.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```
